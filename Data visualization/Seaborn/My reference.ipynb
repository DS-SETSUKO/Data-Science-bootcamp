```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.distplot(tips["total_bill"], bins=30)
```

```python
sns.jointplot(x="total_bill",y="tip", data=tips, kind="hex")
```

```python
sns.jointplot(x="total_bill",y="tip", data=tips, kind="reg")
```

```python
sns.pairplot(tips, hue="sex")
```

```python
sns.rugplot(tips["total_bill"])
```

```python
sns.kdeplot(tips["total_bill"])
```

```python
sns.barplot("sex", "total_bill", data=tips, estimator=np.std)
```

```python
sns.countplot("smoker",data=tips, hue="sex")
```

```python
sns.boxplot("day", "total_bill", data=tips, hue="smoker")
```

```python
sns.violinplot("day", "total_bill", data=tips, hue="smoker", split=True)
```

```python
sns.stripplot("day", "total_bill", data=tips, hue="sex")
```

```python
sns.catplot(x="day", y="total_bill", data=tips, kind="violin")
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc, annot=True, cmap="coolwarm")
```

```python
flights = sns.load_dataset("flights")
```

```python
flights.head()
```

```python
fp = flights.pivot_table("passengers", "month", "year")
```

```python
fp
```

```python
sns.heatmap(fp, cmap="coolwarm", linecolor="white", linewidths=1)
```

```python
sns.clustermap(fp, cmap="coolwarm", standard_scale=1)
```

```python
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_lower(sns.scatterplot)
g.map_upper(sns.kdeplot)
```

```python
g = sns.FacetGrid(data=tips, col="time", row="smoker")
g.map(sns.scatterplot, "total_bill", "tip")
```

```python
sns.lmplot("total_bill", "tip", tips, "sex", markers=["o", "v"], scatter_kws={"s":50})
```

```python
sns.lmplot("total_bill", "tip", tips, "sex", col="day", aspect=.7, height=10)
```

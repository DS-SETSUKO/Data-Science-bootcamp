```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
plt.plot(x,y,"r")
plt.title("Title")
plt.ylabel("Y label")
plt.xlabel("X label")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r")

plt.subplot(1,2,2)
plt.plot(y,x,"b")
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1, 0.8])
axes.plot(x, y)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 1, 0.8])
axes1.set_title("Big plot")
axes1.set_ylabel("Y label")
axes1.set_xlabel("X label")
axes1.plot(x,y,"r")

axes2 = fig.add_axes([0.2, 0.5, 0.5, 0.3])
axes2.set_title("Small plot")
axes2.set_ylabel("Y label")
axes2.set_xlabel("X label")
axes2.plot(y,x,"b")
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y, "r")
axes[1].plot(y,x, "b")

plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1,figsize=(8,5))

axes[0].plot(x,y, "r")
axes[1].plot(y,x, "b")

plt.tight_layout()
```

```python
fig.savefig("plot.png", dpi=200)
```

```python
fig = plt.figure(figsize=(8,4))

axes = fig.add_axes([0,0,1,1])
axes.set_ylabel("Y")
axes.set_xlabel("X")
axes.set_title("Legend")
axes.plot(x, x**2, label="Squared")
axes.plot(x, x**3, label="Cubed")
plt.legend()
```

```python
fig.savefig("legend.png", dpi=200)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y, color="#17a2b8", lw=2, ls="--", marker="o", markersize=10)
```
